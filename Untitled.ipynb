## correlation
In this part, we are making correlation between feature to see how some feature can explain another feature. or see relation between them


```python
#we gonna merge our solar generation data with weather data
p1_weather.head(10)
pdata1.head(10)
power_sensor=pd.merge(p1_weather,pdata1,left_on="DATE_TIME",right_on="DATE_TIME")              
```

```python
power_sensor.info()
```

```python
##we start correlation
power_sensor.corr(method = 'spearman')    
#
```

### DAILY_YIELD is not correlated with all feature but AMBIENT_TEMPERATURE is moreless correlated.TOTAL_YIELD is also not correlated with all feature. I remove it in the correlation matrix.



```python
corr=power_sensor.drop(columns=["DAILY_YIELD","TOTAL_YIELD"]).corr(method="spearman")
```

```python
sna.heatmap(corr,robust=True,annot=True,fmt="0.3f",linewidth=.5,square=True)
plt.show()
```

```python
power_sensor.head()
```

```python
sna.pairplot(power_sensor.drop(columns=["DAILY_YIELD","TOTAL_YIELD"]))
plt.show()
```

```python
#we will plot DC_POWER and AC_POWER
plt.figure(dpi=100)
sna.lmplot(x="DC_POWER",y="AC_POWER",data=power_sensor)
plt.title("regression plot")
plt.show()
```

Comment

    This graph said that inverter convert dc power to ac power linearly.  dcpower=10∗acpower  inverter lost 90% of their power when it convert.

```python
comment

DC_POWER is produced linearly by MODULE_TEMPERATURE with some variability.
```

```python
plt.figure(dpi=100)
sna.lmplot(x='IRRADIATION', y='DC_POWER', data=power_sensor)
plt.title('Regression plot')
plt.show()
```

DC_Power increase with IRRADIATION.

short conclusion

In this section, we conclude that:

Yield does not depend on the Temperature, the dc/ac power and irradiation.

the transfert function between dc and ac power is linear.

dc power is indeed influenced by the ambient temperature, by the temperature of the module, by the irradiation and finally by the heat transfer between the module and the air.
